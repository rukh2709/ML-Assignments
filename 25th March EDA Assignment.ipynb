```python
# Answer 9
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df.shape
```

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.

```python
# Answer 10: 
# Filter the relevant columns
new_df = df[['Category', 'Rating']]

# Group the data by category and calculate the average rating
grouped_df = new_df.groupby('Category')['Rating'].mean().reset_index()
print(grouped_df)
```

```python
# Create a boxplot using matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot([grouped_df[grouped_df['Category'] == category]['Rating'] for category in grouped_df['Category']])
plt.xticks(range(1, len(grouped_df['Category']) + 1), grouped_df['Category'], rotation=90)
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Ratings by App Category in Google Play Store')
plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
#Answer 14:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

# Load the dataset
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Create a frequency table of app installs
install_freq = df.groupby(['App'])['Installs'].sum().reset_index()

# Sort the frequency table by the number of installs in descending order
install_freq_sorted = install_freq.sort_values(by=['Installs'], ascending=False)

# Display the top 10 most popular apps
top_10 = install_freq_sorted.head(10)
print(top_10)
```
